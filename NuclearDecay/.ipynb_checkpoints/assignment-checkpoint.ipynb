```python
#radioactive functions and models required throughout the notebook.

def hl(_lambda):
    '''
    Parameters
    ----------
    _lambda: np.array
        decay constant
    
    Returns
    -------
    half life: np.array
    '''
    return -np.log(0.5)*(_lambda)**-1

def power_law(C, d, alpha):
    '''
    Parameters
    ----------
    C: float
        Constant
    
    d: np.array
        Detector distances
    alpha:
        Power law exponent
        
    Returns
    -------
    A0: np.array
        Initial activity
    '''
    return C * d**alpha

def N0(a0, detector_csa, _lambda, r):
    '''
    Parameters
    ----------
    a0: float
        initial activity
    
    detector_csa: float
        Cross sectional area of the detector
    
    _lambda: float
        Decay constant
    
    r: float
        Detector distance
        
    Returns
    -------
    N0: float
        Initial number of particles
    '''
    return (a0 / detector_csa) * 4 * np.pi * (r)**2 * (_lambda)**-1

def N0_unc(_lambda, e_lambda, r, e_r, Ri, e_Ri, csa, e_csa):
    '''
    Parameters
    ----------
    Ri: float
        initial activity
    e_Ri: float
        Error in initial activity
        
    csa: float
        Cross sectional area of the detector
    e_csa: float
        Error in cross sectional area of the detector
        
    _lambda: float
        Decay constant
    
    e_lambda: float
        Error in decay constant
    
    r: float
        Detector distance
    e_r: float
        Error in detector distance
        
    Returns
    -------
    N0_unc: float
        Error in detector
    '''
    
    sa = 4*np.pi*r**2 # surface area
    a = e_Ri**2 * ((sa/csa)*(_lambda)**-1)**2
    b = e_r**2 * ((Ri/csa)*(4*np.pi*r)*(_lambda)**-1)**2
    c = e_csa**2 * ((Ri/(csa)**2)*(4*np.pi*r**2)*(_lambda**-1))**2
    d = e_lambda**2 *(Ri/csa)*4*np.pi*r**2*(1/_lambda**2)**2
    
    return np.sqrt(a + b + c + d)

def A(t, _lambda, A0):
    '''
    Parameters
    ----------
    t: np.array
        time values
    _lambda: float
        decay constant
    A0: float
        Initial activity
    Returns
    -------
    A: np.array
        Activity series
    '''
    return A0*np.exp(-_lambda*t)

def lnA(t, _lambda, lnA0):
    '''
    Parameters
    ----------
    t: np.array
        time values
    _lambda: float
        decay constant
    lnA0: float
        Log initial activity
    Returns
    -------
    lnA: np.array
        Log of the series parameters
    '''
    return -_lambda*t + lnA0

def linear_power_law(lnD, a, lnC):
    '''
    Parameters
    ----------
    lnC: float
        Constant
    
    lnD: np.array
        Log detector distances
    a:
        Power law exponent
        
    Returns
    -------
    A0: np.array
        Initial activity
    '''
    return a*lnD + lnC

def linear_power_law_cs(params):
    '''
    Helper function for unpacking parameters for injection into functions
    Parameters
    ----------
    params: np.array
        Array of [lnD, a, lnC]
    Returns
    -------
    lnA0 : np.array
        Log of initial activities
    '''
    return params[1]*params[0] + params[2]

def At(t, A01, A02, lambda_1, lambda_2):
    '''
    Parameters
    ----------
    t: np.array
        Time array
    
    A01: constant
        Initial activity value for isotope 1
    
    A02: constant
        Initial activity value for isotope 2
    
    lambda_1: constant
        Decay constant for isotope 1
    
    lambda_2: constant
        Decay constant for isotope 2
    Returns
    -------
    A: np.array
        Activity series
        
    '''
    return A01*np.exp(-lambda_1*t) + A02*np.exp(-lambda_2*t)

def At_cs(params):
    '''
    Helper function of At(), unpacking parameters for injection into functions
    Parameters
    ----------
    params: np.array
        Array of [t, A01, A02, lambda_1, lambda_2]
    Returns
    -------
    A : np.array
        Activitiy series
    '''
    return At(params[0], *params[1:])

def mcmc_At(t, params):
    '''
    Helper function of At(), for mcmc function
    Parameters
    ----------
    t: np.array
        Time array
    params: np.array
        Array of [A01, A02, lambda_1, lambda_2]
    Returns
    -------
    A : np.array
        Activitiy series
    '''
    return At(t, *params)

def LNA(params):
    '''
    Generic linear helper function for fitting params
    '''
    return params[0]*params[1] + params[2]
```

```python
# fitting methods

def least_squares(x, y, error):
    '''
    Parameters
    ----------
    x: np.array
        x value arrays
    y: np.array
        y value arrays
    error: np.array
        error value arrays
    
    Returns
    -------
    (grad, c): tuple
        tuple of gradient and intercept calculated
        with the least squares method
    '''
    
    
    s = np.sum(1/error**2)
    s_x = np.sum(x/error**2)
    s_y = np.sum(y/error**2)
    s_xy = np.sum((x*y)/error**2)
    s_xx = np.sum((x*x)/error**2)
    
    sigma_grad = s/(s*s_xx - (s_x)**2)
    sigma_c = s_xx/(s*s_xx - (s_x)**2)
    
    return (s*s_xy - s_x*s_y)/(s*s_xx - s_x**2), (s_xx*s_y - s_x*s_xy)/(s*s_xx - s_x**2), sigma_grad, sigma_c


def pl_bootstrap(x, y, y_err, num_iterations = 10000):
    '''
    Parameters
    ----------
    
    x:
    y:
    y_err:
    
    num_iterations: int
        number of iterations for the bootstrap method
        
    Returns
    -------
        np.array([grad, unc1_grad, unc2_grad]), (np.array([intercept, unc1_intercept, unc2_intercept]))
    '''
    
    grad, c, _, _ = least_squares(x, y, y_err)

    predict = grad*x + c
    residuals = y - predict        

    grads = []
    cs = []

    for j in range(num_iterations):
        idxs = np.random.randint(residuals.shape[0], size=residuals.shape[0])

        _grad, _intercept, _, _ = least_squares(x, predict + residuals[idxs], y_err[idxs])
        
        grads.append(_grad)
        cs.append(_intercept)

    grads = np.sort(grads)
    cs = np.sort(cs)
    
    return np.array([[np.median(grads), np.percentile(grads, 84) - np.median(grads),  np.median(grads) - np.percentile(grads, 16)], 
                     [np.median(cs), np.percentile(cs, 84) - np.median(cs), np.percentile(cs, 84) - np.median(cs)]])

def chisquare(y, y_predict, y_err):
    return (y_predict - y)**2/y_err**2

def chi2(y_predict, y, y_err):
    return (y_predict - y)**2/y_err**2

def chi2_op(func, x, y, y_err, err, **kwargs):
    '''
    Generic chi2 optimisation function for N dimensional grids
    
    Parameters
    ----------
    func: function
        generic function
    
    x: np.array
    
    y: np.array
    
    y_err: np.array
        Errors in y values
    
    err: np.array
        Parameter error, ie 2.3 for two paramters, 4.72 for four parameters
    
    **kwargs
    Series of keyword arguments. Set of linspaces that the grid is created from.
    
    Returns
    -------
    chi2_min, best fit parameter values and the uncertainties in the parameters
    
    '''
    
    *grid, _x  = np.meshgrid(*list(kwargs.values()), x)

    chi_grid = np.sum((func([_x, *grid]) - y)**2/y_err**2, axis = len(kwargs))
    
    idx = np.unravel_index(np.argmin(chi_grid, axis = None), chi_grid.shape)
        
    grid_idxs = np.where(chi_grid - np.min(chi_grid) <= err)
    
    best_vals = np.array([list(kwargs.values())[i][idx[i]] for i in range(len(idx))])
    
    unc_best_vals = np.array([np.max(abs(best_vals[i] - (list(kwargs.values())[i])[grid_idxs[i]])) for i in range(len(idx))])
    
    
    return np.min(chi_grid), best_vals , unc_best_vals

def AIC(chi_square: float, p: float):
    '''
    Aikiake information criterion
    
    Parameters
    ----------
    
    chi_square: float
        Chi square value
    p: int
        Number of parameters
    
    Returns
    -------
    AIC: float
    '''
    return chi_square + 2*p

def BIC(chi_square: float, p: float, N: int):
    '''
    Bayesian information criterion
    
    Parameters
    ----------
    
    chi_square: float
        Chi square value
    p: int
        Number of parameters
    N: int
     Number of data points
    
    Returns
    -------
    BIC: float
    '''
    
    return chi_square + np.log(N)*p

def mcmc(func,x,y,sig,pars0,stepsize,nstep=1e4):
    '''
    mcmc:  function to perform a simple MCMC using the Metropolis-Hastings Algorithm

    Inputs:
    func  - the function to be used for fitting. Function should accept 2 inputs
            the first is the x-values at which the model is calculated, the second
            is a numpy array for the parameters.
    x     - the x-values at which the model should be calculated (numpy array)
    y     - the value of the data for each x (numpy array)
    sig   - the uncertainty on y (numpy array)
    pars0 - The initial paramters from which to start the MCMC chain (numpy array)
    stepsize - the stepsize for each of the parameters (numpy array)
    nstep    - the number of steps in the MCMC chain

    Outputs:
    chain    - The MCMC chain as a (Nsteps, Npars) numpy array
    '''
    nstep=int(nstep)
    npars=pars0.shape[0]
    chain=np.zeros((int(nstep),npars))
    chi2=np.zeros(int(nstep))

    chain[0,:]=pars0.copy() 
    mdl=func(x, pars0) #1
    chi2[0]=np.sum( (y-mdl)**2/sig**2) #2
    njump=0
    for i in range(1,nstep):
        pars_new=abs(np.random.normal(chain[i-1,:],stepsize)) #4
        
        mdl=func(x,pars_new) #5
        chi2_new=np.sum( (y-mdl)**2/sig**2) #6
        if (chi2_new > chi2[i-1]):
            p0=np.exp(-(chi2_new-chi2[i-1])/2.) # 7
            p=np.random.uniform(0,1,1)
            if  p<=p0: # 8
                chain[i,:]=pars_new
                chi2[i]=chi2_new
                njump=njump+1
            else:
                chain[i,:]=chain[i-1,:]
                chi2[i]=chi2[i-1]
        else:
            chain[i,:]=pars_new
            njump=njump+1
            chi2[i]=chi2_new
    print('Jump fraction: %.3f' %(njump/nstep))
    return np.insert(chain, chain.shape[1] ,chi2, axis = 1)
```

```python
def weighted_mean(x, sigma):
    '''
    Weighted mean of series of measurements
    
    Parameters
    ----------
    x: np.array
        Measurements
    sigma: np.array
        Uncertainties in measurements
    
    Returns
    -------
    Weighted Mean: float
    
    '''
    return np.sum(x/sigma**2)/np.sum(sigma**-2)
def weighted_error(sigma):
    '''
    Error in the weighted mean
    Parameters
    ----------

    sigma: np.array
        Uncertainties in measurements
    
    Returns
    -------
    Error: float
    
    '''
    return np.sqrt(np.sum(sigma**-2)**-1)
```

```python
# plotting
def plot(x, y, yerr, y_predict, xlabel="", ylabel="", fit = True):
    '''
    Generic plot function
    '''
    fig = plt.figure()
    
    plt.errorbar(x, y, yerr = yerr, c="black", fmt="x")
    
    if fit:
        plt.plot(x, y_predict, c="red")
    plt.xlabel(xlabel, fontsize=18)
    plt.ylabel(ylabel, fontsize=18)
    return fig
        
def plot_with_residuals(x, y, yerr, y_predict, fit = True):
    '''
    Plot function with residuals
    
    '''
    plt.rc('font', size=15) 
    fig1 = plt.figure(figsize=(8,8))

    frame1=fig1.add_axes((.1,.3,.8,.6))


    plt.errorbar(x, y, yerr = yerr, c="black", fmt="x")
    
    if fit:
        plt.plot(x, y_predict, c="red")


    frame1.set_xticklabels([])
    frame1.set_ylabel(r"A / $s^{-1}$", fontsize=20)

    frame2=fig1.add_axes((.1,.08,.8,.2))    

    frame2.set_ylabel(r"$\delta$ / $s^{-1}$", fontsize=20)
    frame2.set_xlabel("Time / s", fontsize=18)
    plt.errorbar(x, y_predict - y, yerr=yerr, c="black", fmt="x")
    #plt.show()
    return fig1
```

```python
def part_2_3(i, params, stepsize, nstep):  
    '''
    Helper function in this part to be called in generator below
    Will output chains and correlation plots
    '''
    reps = 5
    print(f"Detector {i}")
    x = np.concatenate(
        [(mcmc(mcmc_At, x = data_2['Time'], y = data_2[f'A{i}'], sig = data_2[f'e_A{i}'], pars0=np.array([np.random.normal(params[i], 10*stepsize[i]) for i, _ in enumerate(params)]), stepsize = stepsize, nstep=nstep)[int(0.1*nstep):]) for j in range(reps)]
    )
        
    data = {r'$A_{0(1)} / s^{-1}$': x[:, 0], r'$A_{0(2)} / s^{-1}$': x[:, 1], r'$\lambda_1 / s^{-1}$': x[:, 2], r'$\lambda_2 / s^{-1}$': x[:, 3]}

    sns.set_theme(style="ticks")

    df = pd.DataFrame(data)

    pplot = sns.pairplot(df, corner=True)
    #pplot.savefig(f"correlation_detector_{i}.png") # creates correlation plot
    plt.show()

    vals = np.array([(np.median(x[:, param]), np.percentile(x[:, param], 84), np.percentile(x[:, param], 16)) for param in range(len(params))])
        
    print("------")
    
    return np.min(x[:, -1]), vals
```

```python
# run the MCMC function for each detector, save to numpy object for use later.
#Note for time efficiency nstep is set to 1e3, for the report 2e5 was used.

nstep = 1e3 #2e5 
part2_4_mcmc = np.array([
    part_2_3(i + 1, params, stepsizes[i], nstep = nstep) for i, params in enumerate(cs_params)
], dtype=object)
```

```python
for i, vals in enumerate(part2_4_mcmc):
    print(f"Detector {i + 1}")
    print("-------")
    fig = plot_with_residuals(
        x = data_2['Time'],
        y = data_2[f'A{i + 1}'],
        yerr = data_2[f'e_A{i + 1}'],
        y_predict = At(data_2['Time'], *vals[1][:, 0])
    )
    #fig.savefig(f"part2_3_detector{i + 1}.png", bbox_inches = "tight")
    plt.show()
    print(f"Chi2 : {vals[0]}")
    print(AIC(vals[0], p = 4), BIC(vals[0], p = 4, N = len(data_2['Time'])))
    
    for i, param in enumerate(vals[1]):
        if i == 0 or i == 1:
            print(f"A0{i + 1} is {param[0]} +/- {abs(param[0] - np.max(param[1:]))}")
        else:
            print(f"lambda{i + 1} is {param[0]} +/- {abs(param[0] - np.max(param[1:]))}")
            #print(hl(param[0]), hl(abs(param[0] + param[1])), hl(abs(param[0] - param[2])))
            print(hl(param[0]), hl(param[0]) - hl(param[1]), hl(param[2]) - hl(param[0]))
    print("-------")
```
